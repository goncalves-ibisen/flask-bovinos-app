```python
import pandas as pd
import json
import datetime
import math
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
dados = pd.read_csv("dados_animais.csv")
```

```python
# Carregar o valor do arquivo JSON
with open('valor_boi_hoje.json', 'r') as file:
    data = json.load(file)
    valor_boi = float(data['valor'].replace(',', '.'))  # Convertendo para float

# Supondo que seu DataFrame se chame 'dados'
# Verificar se a coluna 'valor' existe e, se sim, removê-la
if 'valor' in dados.columns:
    dados = dados.drop(columns=['valor'])

# Criar a nova coluna 'valor' multiplicando a coluna 'arrobas' pelo valor do arquivo JSON
dados['valor'] = dados['arrobas'] * valor_boi
```

```python
# Definindo o intervalo de datas
data_inicio = pd.to_datetime('2020-01-01')
data_atual = pd.to_datetime(datetime.now().strftime('%Y-%m-%d'))

# Função para gerar uma data aleatória no intervalo definido
def gerar_data_aleatoria(inicio, fim):
    return inicio + pd.to_timedelta(np.random.randint(0, (fim - inicio).days + 1), unit='D')

# Verificar se a coluna 'nascimento' existe e, se sim, removê-la
if 'nascimento' in dados.columns:
    dados = dados.drop(columns=['nascimento'])

# Criar a nova coluna 'nascimento' com datas aleatórias no intervalo
dados['nascimento'] = [gerar_data_aleatoria(data_inicio, data_atual).strftime('%d-%m-%Y') for _ in range(len(dados))]
```

```python
# Supondo que seu DataFrame se chame 'dados' e tenha uma coluna 'nascimento' no formato 'dia-mes-ano'
# Vamos garantir que a coluna 'nascimento' esteja no formato de data
dados['nascimento'] = pd.to_datetime(dados['nascimento'], format='%d-%m-%Y')

# Função para calcular a idade em meses
def calcular_idade_em_meses(data_nascimento):
    hoje = datetime.now()
    idade_meses = (hoje.year - data_nascimento.year) * 12 + (hoje.month - data_nascimento.month)
    return idade_meses

# Adicionando a coluna 'idade_meses' ao DataFrame
dados['idade_meses'] = dados['nascimento'].apply(calcular_idade_em_meses)

# Função para classificar os bovinos
def classificar_bovino(row):
    if 0 < row['idade_meses'] <= 12:
        return 'Bezerro'
    elif 12 < row['idade_meses'] <= 24:
        return 'Garrote'
    elif 24 < row['idade_meses'] <= 36:
        return 'Novilho'
    return 'Boi'  # Caso não seja Macho ou não se encaixe em nenhuma classificação

# Adicionando a coluna 'classificacao' ao DataFrame
dados['classificacao'] = dados.apply(classificar_bovino, axis=1)

# Exibindo as primeiras linhas para verificar o resultado
print(dados.head())
```

```python
# Função para atribuir valores aleatórios à coluna 'arrobas' com base na classificação
def calcular_arrobas(classificacao):
    if classificacao == 'Bezerro':
        return np.random.randint(2, 11)  # Valor entre 2 e 10
    elif classificacao == 'Garrote':
        return np.random.randint(10, 20)  # Valor entre 10 e 19
    elif classificacao == 'Novilho':
        return np.random.randint(20, 31)  # Valor entre 20 e 30
    else:
        return np.random.randint(31, 47)  # Valor entre 31 e 46 para outras classificações

# Aplicar a função para preencher a coluna 'arrobas' de acordo com a classificação
dados['arrobas'] = dados['classificacao'].apply(calcular_arrobas)
```

```python
def gerar_grafico_barras(num_machos, num_femeas):
    # Definir dados para o gráfico
    categorias = ['Machos', 'Fêmeas']
    valores = [num_machos, num_femeas]

    # Criar o gráfico de barras
    plt.figure(figsize=(6, 4))
    plt.bar(categorias, valores, color=['blue', 'pink'])
    plt.xlabel('Sexo')
    plt.ylabel('Quantidade')
    plt.title('Comparação de Machos e Fêmeas')

    # Salvar o gráfico como uma imagem na pasta 'static'
    # plt.savefig(r'C:\cardim sistemas\static\grafico_machos_femeas.png')
    plt.savefig('static/grafico_machos_femeas.png')
    plt.close()
```

```python
def carregar_dados_hist_preco():
    # Coloque aqui a lógica para carregar os dados
    df = pd.read_csv('dados/dados-historico-boi-gordo.csv', delimiter=';')
    df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y', errors='coerce')
    df['valor-rs'] = pd.to_numeric(df['valor-rs'].str.replace(',', '.'), errors='coerce')
    return df
```

```python
dados_preco = carregar_dados_hist_preco()

# Filtrar os dados para os últimos 30 dias
last_30_days = datetime.now() - timedelta(days=30)
df_last_30_days = dados_preco[dados_preco['data'] >= last_30_days]
df_last_30_days = df_last_30_days.sort_values(by='data')

# Gerar o gráfico e salvar como imagem
plt.figure(figsize=(6, 4))  # Ajuste para um tamanho máximo de 420px
plt.bar(df_last_30_days['data'].dt.strftime('%Y-%m-%d'), df_last_30_days['valor-rs'])
plt.xlabel('Data')
plt.ylabel('Valor (R$)')
plt.title('Valores do Boi Gordo nos Últimos 30 Dias')
plt.xticks(rotation=45)
plt.tight_layout()
image_path = 'static/images/grafico_boi_gordo.png'
plt.savefig(image_path)
```
